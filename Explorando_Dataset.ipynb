## Criando um dataframe de personagens de Star Wars

```python
personagens_df = pd.DataFrame({
     'nome':['Luke Skywalker','Yoda','Palpatine'],
     'idade':[20,1000,70],
     'peso':[60.1,15.2,70.5],
     'é jedi?':[True,True,False]
    })
personagens_df.info()
personagens_df.columns
```

Renomeando Colunas de um DataFrame com DataFrame.rename(columns)

```python
# Para renomear uma coluna de um df voce pode usar DataFrame.rename
# que cria uma copia com as colunas renomeadas
# temos que passar o atributo columns para renomear uma coluna
personagens_df_renomeado = personagens_df.rename(columns={
      'nome': 'Nome Completo',
      'idade': 'Idade'})
personagens_df_renomeado
```

Para renomear o DF original precisamos passar o parametro `inplace=True`


```python
personagens_df.rename(columns={
    'nome':'Nome Completo',
    'idade':'Idade'
},inplace=True)

personagens_df
```

Tambem podemos renomear todas as colunas de um DF é utilizar o atributo `DataFrame.columns `passando uma lista com os novos nomes

```python

personagens_df.columns = ['NOME','IDADE','PESO','É JEDI']
personagens_df
```
