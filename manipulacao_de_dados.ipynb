```python
import pandas as pd

df = pd.read_csv('2004-2021.tsv', sep='\t')
df
```

```python
# adicionando coluna com valor default
df['Coluna sem nocao'] = 'DEFAULT'
df

# adicionando coluna a partir de lista
df['Coluna a partir de lista'] = range(df.shape[0])
df

# adicionando coluna com base em operacao em outra coluna
df['preco medio revenda (dolar)'] = df['PREÇO MÉDIO REVENDA'] * 6
df
```
